```python
a["mean score"]=((a["math score"]+a["reading score"]+a["writing score"])/3) .round()
a.head(20)
```

```python
#Label Encoding
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
a['gender']=le.fit_transform(a['gender'])
a['race/ethnicity']=le.fit_transform(a['race/ethnicity'])
a['parental level of education']=le.fit_transform(a['parental level of education'])
a['lunch']=le.fit_transform(a['lunch'])
a['test preparation course']=le.fit_transform(a['test preparation course'])
a.head(20)
```

```python
#data Processing
a=a.drop(['math score','writing score','reading score'],axis=1)
a.head()
```

```python
from sklearn.model_selection import train_test_split
x=a.drop('mean score',axis=1)
y=a['mean score']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression

# Create the Logistic Regression model
model = LogisticRegression(solver='liblinear', random_state=0)

# Fit the model to the training data
model.fit(x_train, y_train)
```

```python
predictions=model.predict(x_test)
predictions
```

```python
y_test
```

```python
difference=abs(predictions-y_test)
difference
```
